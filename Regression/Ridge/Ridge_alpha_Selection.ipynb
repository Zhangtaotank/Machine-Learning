```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
x = 1. / (np.arange(1,11)+np.arange(0,10)[:, np.newaxis])
y = np.ones(10)
```

```python
n_alphas = 200
alphas = np.logspace(-10,-2,n_alphas)
```

```python
coef = []
for a in alphas:
    ridge=linear_model.Ridge(alpha=a, fit_intercept=False)
    ridge.fit(x,y)
    coef.append(ridge.coef_)
```

```python
len(coef)
```

```python
%matplotlib inline
```

```python
ax = plt.gca() # "GetCurrentAxis"
ax.plot(alphas, coef)
plt.show()
```

```python
ax = plt.gca() # "GetCurrentAxis"
ax.plot(alphas, coef)
ax.set_xscale("log")
plt.show()
```

```python
ax = plt.gca() # "GetCurrentAxis"
ax.plot(alphas, coef)
ax.set_xscale("log")
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.show()
```

```python
ax = plt.gca()

ax.plot(alphas, coef)
ax.set_xscale('log')
ax.set_xlim(ax.get_xlim()[::-1])  # reverse axis
plt.xlabel('alpha')
plt.ylabel('weights')
plt.title('Ridge coefficients as a function of the regularization')
plt.axis('tight')
plt.show()
```
